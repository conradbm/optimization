```python
import random


class Points():
    def __init__(self, x1,x2,d1,d2,z):
        self.x1=x1
        self.x2=x2
        self.d1=d1
        self.d2=d2
        self.z=z
```

```python
def initfibonicciarray (fibo,N):
     fibo.append(0)
     fibo.append(1)
     for i in range(2,N):
         fibo.append(fibo[i-1]+fibo[i-2])
     


def fiboniccisearch(points,i,N, fibo):
    lamlo=0
    lamhi=1
    for k in range(0,N-2):
        a= fibo[N-k-3]/fibo[N-k-1]*(lamhi-lamlo)
      #  print a, fibo[N-k-3],fibo[N-k-2], N-k-3, float (fibo[N-k-3])/float(fibo[N-k-1])
        lammid1=lamlo+float( fibo[N-k-3])/float (fibo[N-k-1])*(lamhi-lamlo)
        zmid1=getobjectivelamd (points,i,lammid1)
        lammid2=lamlo+float (fibo[N-k-2])/float (fibo[N-k-1])*(lamhi-lamlo)
        zmid2=getobjectivelamd (points,i,lammid2)
        # print k, lamlo, lammid1, lammid2, lamhi, 'z ',zmid1, zmid2
        # raw_input()
        if zmid1<zmid2:
            lamhi=lammid2
        else:
            lamlo=lammid1
    return (lammid1)

def golden_search(points,i,low=0,high=1,eps=0.1, verbose=False):

    # Manually plug in based on problem
    #min (x1-2)^4 +(x1-2x2)^2
    fun = lambda l: ((points[i].x1-l*points[i].d1)-2)**4 + ((points[i].x1-l*points[i].d1)-2*((points[i].x2-l*points[i].d2)))**2 
    #fun = lambda l:  (1/2)*(xi[0]-l*d[0])**2 + (xi[0]-l*d[0])*(xi[1]-l*d[1])+ (1/2)*(xi[1]-l*d[1])**2
    #fun = lambda l: (xi[0]-l*d[0])*(xi[1]-l*d[1])
    if verbose:
        print("*******************************")
        print("Executing Golden Ratio Search")
        print("*******************************")
        
        
    
    # Define Constants
    alpha = 0.618
    k=0
    
    # Create lists
    lam=list()
    mu=list()
    a=list()
    b=list()
    distances=list()
    
    # Step 1
    a.append(low)
    b.append(high)

    lam0 = a[k] + (1-alpha)*(b[k]-a[k])
    mu0 = a[k] + alpha*(b[k]-a[k])
    
    lam.append(lam0)
    mu.append(mu0)
    
    if verbose:
        print("************************")
        print(("[a,b]"), "|b[k]-a[k]|")
        print("************************")
        
    while True:
        distances.append(abs(b[k]-a[k])) 
        if verbose:
            print(a[k],b[k], distances[k])

        if(distances[k] < eps): # optimal soln lies within [a,b]
            break
        elif fun(lam[k]) > fun(mu[k]):
            # Step 2
            a.append(lam[k])
            b.append(b[k])
            lam.append(mu[k])
            mu.append(a[k+1] + alpha*(b[k+1]-a[k+1]))
        elif fun(lam[k]) <= fun(mu[k]):
            # Step3
            a.append(a[k])
            b.append(mu[k])
            mu.append(lam[k])
            lam.append(a[k+1]+(1-alpha)*(b[k+1]-a[k+1]))
        else:
            print("Something went wrong.")

        k += 1
    
    #return random.uniform(a[-1], b[-1])
    return (a[-1]+b[-1])/2

def getobjectivelamd (points, i,lambd):
    z=pow(points[i].x1-lambd*points[i].d1-2,4)+pow(points[i].x1-lambd*points[i].d1-2*points[i].x2+lambd*points[i].d2, 2)
    return(z)

def getobjective (x1,x2):
    z=pow(x1-2,4)+pow(x1-2*x2, 2)
    return(z)

def getgradient1 (x1,x2):
    d1=4*pow(x1-2,3)+2*(x1-2*x2)
    return(d1)

def getgradient2 (x1,x2):
    d2=-4*(x1-2*x2)
    return(d2)
```

```python

#MAIN


points = []
fp=open('output.txt','w')
fp.close()

fibo=[]
N=40
initfibonicciarray (fibo,N)
#print fibo
#min (x1-2)^4 +(x1-2x2)^2
iteration=0
x1=0
x2=3
z=getobjective(x1,x2)
for i in range (0,100):
    d1=getgradient1(x1,x2)
    d2=getgradient2(x1,x2)
    points.append(Points(x1,x2,d1,d2,z))
    lambd= golden_search(points,i,verbose=True)
    print   ('z =' , z,'x=(',x1,',',x2,') d=(',d1,',', d2,')', lambd)
    x1=x1-lambd*d1
    x2=x2-lambd*d2
    z=getobjective(x1,x2)


print ('FINAL SOLUTION x=(',x1,',',x2,')  z =' , z)
```
