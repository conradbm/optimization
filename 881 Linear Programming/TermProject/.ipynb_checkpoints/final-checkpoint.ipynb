# Problem 1:

Reproduce example in Lecture 10 notes using Primal Affine Scaling Algorithm

```python
A = np.matrix([[1, -1, 1, 0],
               [0, 1 , 0, 1]])
b = np.array([15, 15])
c = np.array([-2, 1, 0, 0])
N = A.shape[0]
M = A.shape[1]
print("A.shape{}".format(A))
print("b.shape {}".format(b))
print("c.shape {}".format(c))
print("{} constraints".format(N))
print("{} dimensions".format(M))

```

```python
def get_dual(X):
    w = np.linalg.inv(A.dot(X**2).dot(A.T)).dot(A).dot(X**2).dot(c)
    s = c - A.T.dot(w.T).T
    return w, s

def get_duality_gap(x,s):
    return x.dot(s.T)

def interior_point(A0,b0,c0,M0,etol=0.0001,alpha=0.99):
    
    # Get IBFS
    x0 = np.random.randn(M0)
    
    # Diagonalize
    X0 = np.diag(x0)
    
    # Get dual variables and dual slack
    w0, s0 = get_dual(X0)
    
    # Get initial duality gap
    gap = get_duality_gap(x0, s0).tolist()[0][0]
    it = 0
    gaps = []
    iters = []
    zs = []
    while True:

        # Get next feasible improving direction
        dy = -(X0**2).dot(s0.T)

        # Get next step
        #if (dy >= 0).all():
        #    alpha = alpha
        #else:
        #    alpha = 1/-min(dy)
        x0 = x0 + alpha*X0.dot(dy).T

        # Get dual slack
        X0 = np.diag(np.asarray(x0)[0])
        w0, s0 = get_dual(X0)

        # Get duality gap
        gap = get_duality_gap(x0, s0).tolist()[0][0]
        it += 1
        z = x0.dot(c)
        
        # Record data
        gaps.append(gap)
        iters.append(it)
        zs.append(z)
           
        print("   iteration {}\n   gap {}\n   x0 {}\n   w0 {}\n   s0 {}\n   alpha {}\n   dy {}\n".format(it, gap, x0, w0, s0, alpha, dy))

        if np.allclose(A.dot(x0.T), b) and (dy > 0).all():
            print("UNBOUNDED EXITING")
            break
        
        if np.allclose(A.dot(x0.T), b) and np.allclose(dy, np.zeros(len(dy))):
            print("OPTIMAL EXITING")
            break

    print("Final Duality Gap {}".format(gap))
    print("Primal Variables {}".format(x0))
    print("Dual Slack Variables {}".format(s0))
    print("Dual Variables {}".format(w0))
    return x0,w0,s0,gaps,iters,zs


```

```python
x0,w0,s0,gaps,iters, zs=interior_point(A,b,c,M,etol=0.001)
```

```python
plt.scatter(iters,gaps)
plt.title("Lecture 10 Problem | Duality Gap")
plt.xlabel("Iteration")
plt.ylabel("Gap")
plt.show()
```

```python
plt.scatter(iters,zs)
plt.title("Lecture 10 Problem | Objective Function")
plt.xlabel("Iteration")
plt.ylabel("z")
plt.show()
```
